```python
import pandas as pd
import requests
```

```python
reddit_api_file = '/content/drive/MyDrive/.api_keys/reddit_api.csv'
api_df = pd.read_csv(reddit_api_file)
client = api_df['CLIENT_ID'].values[0]
secret = api_df['KEY'].values[0]
password = api_df['PASSWORD'].values[0]
```

```python
auth = requests.auth.HTTPBasicAuth(client, secret)
```

```python
data = {'grant_type': 'password',
        'username': 'GattoP666',
        'password': password}

auth
```

```python
headers = {'User-Agent': 'UdemyTutorial/0.0.1'}
res = requests.post('http://www.reddit.com/api/v1/access_token', auth=auth, data=data, headers=headers)
res.content
```
